```python
import gprof_log_manipulation as gp
import os
from os import path, listdir
from pprint import pprint
import pandas as pd

p = 'AI/'

files = [f for f in listdir(p) if path.isfile(path.join(p, f))]

d_dict = {}

b = gp.gprofLogManip(p+files[0])
b = b.returnDict()
for key in b.keys():
    d_dict[key] = []

for f in files:
    b = gp.gprofLogManip(p+f)
    b = b.returnInfo()
    for key in b.keys():
        d_dict[key].append(b[key])

print(pd.DataFrame(d_dict))


```
